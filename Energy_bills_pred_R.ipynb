# K-Means clustering 1

image.png

Elbow method

```r
wcss <- numeric()

for (k in 1:10) {
  kmeans_result <- kmeans(df[, c("housearea", "ave_monthly_income", "amount_paid")], centers = k)
  wcss[k] <- sum(kmeans_result$withinss)
}

plot(wcss, type = "o", pch = 19, col = "blue",
     xlab = "Number of Clusters (k)", ylab = "WCSS",
     main = "Elbow Method")
```

K-Means 3 clusters

```r
df_kmeans <- kmeans(df[, c("housearea", "ave_monthly_income", "amount_paid")], centers = 3)

# for visualization
df_clustered <- df # clone df
df_clustered$cluster = as.factor(df_kmeans$cluster) # add cluster column
```

```r
ggplot(df_clustered, aes(x = housearea, y = amount_paid, color = cluster)) + geom_point() +
  labs(title = "K-means Clustering", x = "House area", y = "Amount paid") +
    theme_light()
```

```r
ggplot(df_clustered, aes(x = housearea, y = ave_monthly_income, color = cluster)) + geom_point() +
  labs(title = "K-means Clustering", x = "House area", y = "Average monthly income") +
  theme_light()
```

```r
ggplot(df_clustered, aes(x = ave_monthly_income, y = amount_paid, color = cluster)) + geom_point() +
  labs(title = "K-means Clustering", x = "Average monthly income", y = "Amount paid") +
    theme_light()
```

**Ở biểu đồ thứ nhất, tiền điện được chia thành 3 cụm rõ ràng, các nhà có diện tích trong khoảng 500 đến 1000 có tiền điện cao hơn so với nhà diện tích dưới 500 và trên 1000. Biểu đồ thứ hai cho thấy diện tích nhà không có nhiều tương quan với thu nhập. Biểu đồ thứ ba chia thu nhập thành 3 cụm như biểu đồ trước, nhà có thu nhập từ 20000 đến dưới 40000 có tiền điện thấp hơn so với các nhà khác.**

# Linear regression

Load train data

```r
lr_train <- df[1:900, ] # load 90% dataset
```

Training

```r
lr_model <- lm(amount_paid ~ num_people + housearea + is_ac + is_tv + is_flat + ave_monthly_income + num_children + is_urban, data = lr_train)
summary(lr_model)
```

Load test data

```r
lr_test <- df[901:1000, 1:10] # load the left 10%
```

Testing

```r
predictions <- predict(lr_model, lr_test)
lr_test$prediction <- predictions

lr_test
```

**Mô hình giải thích được 88% sự biến động của amount_paid với chỉ số multiple R-squared bằng 0.8808. Sai số cao nhất so với thực tế là -132.293. Thuộc tính is_urban có hệ số cao nhất chứng tỏ việc nhà ở thành thị hay nông thôn sẽ ảnh hưởng lớn nhất đến tiền điện, ngược lại diện tích nhà không ảnh hưởng nhiều đến tiền điện nhưng vẫn có ý nghĩa thống kê đối với mô hình.**
